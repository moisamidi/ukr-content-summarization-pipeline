```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, pipeline
import torch
```

```python
models = {
    "UA_NEWS_T5": "SGaleshchuk/t5-large-ua-news",
    "MT5_XLSUM": "csebuetnlp/mT5_multilingual_XLSum",
    "UK_SUMMARIZER": "ukr-models/uk-summarizer"
}
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
```

```python
with open("chunks.txt", "r", encoding="utf-8") as f:
    chunks = [c.strip() for c in f.read().split("\n\n") if c.strip()]

print("Chunks count:", len(chunks))
```

# Сontrolled Comparison of Summarization Models

```python
generation_params = {
    "max_new_tokens": 256,
    "min_new_tokens": 80,
    "num_beams": 4,
    "no_repeat_ngram_size": 3,
    "repetition_penalty": 1.1,
    "length_penalty": 1.1,
    "early_stopping": True,
    "do_sample": False
}
```

```python
def run_summarization_experiment(models, chunks, generation_params, output_file_path):
    output_file = open(output_file_path, "w", encoding="utf-8")

    for model_name, model_path in models.items():
        print(f"\n\n========== MODEL: {model_name} ==========\n")
        output_file.write(f"\n\n========== MODEL: {model_name} ==========\n\n")

        tokenizer = AutoTokenizer.from_pretrained(model_path)
        model = AutoModelForSeq2SeqLM.from_pretrained(model_path)
        model.to(device)

        for i, chunk in enumerate(chunks, start=1):
            inputs = tokenizer(chunk, return_tensors="pt").to(device)
            input_tokens_count = inputs.input_ids.shape[1]

            outputs = model.generate(**inputs, **generation_params)

            summary = tokenizer.decode(outputs[0], skip_special_tokens=True)
            output_tokens_count = len(tokenizer.encode(summary, add_special_tokens=False))

            print(f"\n--- CHUNK {i} ---\n")
            print(f"ORIGINAL ({input_tokens_count} tokens):\n")
            print(chunk)
            print(f"\nSUMMARY ({output_tokens_count} tokens):\n")
            print(summary)
            print("\n" + "=" * 60)

            output_file.write(f"\n--- CHUNK {i} ---\n")
            output_file.write(f"\nORIGINAL ({input_tokens_count} tokens):\n")
            output_file.write(chunk + "\n")
            output_file.write(f"\nSUMMARY ({output_tokens_count} tokens):\n")
            output_file.write(summary + "\n")
            output_file.write("\n" + "=" * 60 + "\n")

    output_file.close()

    print(f"\nDone. Results saved to {output_file_path}")
```

```python
run_summarization_experiment(
    models=models,
    chunks=chunks,
    generation_params=generation_params,
    output_file_path="controlled_comparison.txt"
)
```

## Model-Specific Evaluation



```python
def run_model_specific_experiment(models, chunks, output_file_path):
    output_file = open(output_file_path, "w", encoding="utf-8")

    model_params = {
        "UA_NEWS_T5": {
            "max_new_tokens": 120,
            "min_new_tokens": 50,
            "num_beams": 4,
            "no_repeat_ngram_size": 3,
            "repetition_penalty": 1.2,
            "early_stopping": True,
            "do_sample": False
        },
        "MT5_XLSUM": {
            "max_new_tokens": 100,
            "min_new_tokens": 50,
            "num_beams": 4,
            "no_repeat_ngram_size": 3,
            "repetition_penalty": 1.2,
            "early_stopping": True,
            "do_sample": False
        },
        "UK_SUMMARIZER": {
            "max_new_tokens": 100,
            "min_new_tokens": 50,
            "num_beams": 4,
            "no_repeat_ngram_size": 3,
            "repetition_penalty": 1.1,
            "early_stopping": True,
            "do_sample": False
        }
    }

    for model_name, model_path in models.items():
        print(f"\n\n========== MODEL: {model_name} ==========\n")
        output_file.write(f"\n\n========== MODEL: {model_name} ==========\n\n")

        tokenizer = AutoTokenizer.from_pretrained(model_path)
        model = AutoModelForSeq2SeqLM.from_pretrained(model_path)
        model.to(device)

        gen_params = model_params[model_name]

        for i, chunk in enumerate(chunks, start=1):
            input_text = "summarize: " + chunk

            inputs = tokenizer(
                input_text,
                return_tensors="pt",
                truncation=True,
                max_length=512
            ).to(device)

            input_tokens_count = inputs.input_ids.shape[1]

            outputs = model.generate(**inputs, **gen_params)

            summary = tokenizer.decode(outputs[0], skip_special_tokens=True)
            output_tokens_count = len(tokenizer.encode(summary, add_special_tokens=False))

            print(f"\n--- CHUNK {i} ---\n")
            print(f"ORIGINAL ({input_tokens_count} tokens):\n")
            print(chunk)
            print(f"\nSUMMARY ({output_tokens_count} tokens):\n")
            print(summary)
            print("\n" + "=" * 60)

            output_file.write(f"\n--- CHUNK {i} ---\n")
            output_file.write(f"\nORIGINAL ({input_tokens_count} tokens):\n")
            output_file.write(chunk + "\n")
            output_file.write(f"\nSUMMARY ({output_tokens_count} tokens):\n")
            output_file.write(summary + "\n")
            output_file.write("\n" + "=" * 60 + "\n")

    output_file.close()
    print(f"\nDone. Results saved to {output_file_path}")
```

```python
run_model_specific_experiment(models, chunks, "model_specific_results.txt")
```
